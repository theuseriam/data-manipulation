## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np

Baltimore = pd.read_csv('baltimore.csv')
New_York = pd.read_csv('new-york.csv')
Boston = pd.read_csv('boston.csv')
Philadelphia = pd.read_csv('philadelphia.csv')
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
for col in Boston.columns:
    pct_missing = np.mean(Boston[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
for col in Baltimore.columns:
    pct_missing = np.mean(Baltimore[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
    
    
#delete the columns with no data
Baltimore = Baltimore.drop(["Type", "Grade", "Origin District", "Color", "Environment", "Quality", "Condition", "Appearance","Storage", "Crop", "Trans Mode", "Sub Variety"], axis=1)

Baltimore.head()
```

```python
for col in New_York.columns:
    pct_missing = np.mean(New_York[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))

    
#delete the columns with no data
New_York = New_York.drop(["Type", "Grade", "Origin District", "Color", "Environment", "Quality", "Condition", "Appearance","Storage", "Crop", "Trans Mode", "Sub Variety"], axis=1)

New_York.head()
```

```python
for col in Philadelphia.columns:
    pct_missing = np.mean(Philadelphia[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
#delete the columns with no data
Philadelphia = Philadelphia.drop(["Type", "Grade", "Origin District", "Color", "Environment", "Quality", "Condition", "Appearance","Storage", "Crop", "Trans Mode", "Sub Variety"], axis=1)

Philadelphia.head()
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
Northeast_Region = pd.concat([Philadelphia,Baltimore, Boston, New_York],
                            keys = ['Philadelphia', 'Baltimore', 'Boston', 'New York'])
Northeast_Region
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.

low_price = Northeast_Region.groupby('Low Price')
high_price = Northeast_Region.groupby('High Price')
low_price.first()
high_price.first()
low_price.agg('mean')
high_price.agg('mean')
```
